```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import nafot
from tqdm import tqdm, tqdm_notebook
import os
```

Find unique stat areas in cellular data

```python
file_names = os.listdir('M:Dor/new samples/with_stat_area')

stat_area_ids = []
for file_name in file_names:
    cur_df = pd.read_csv(f'M:Dor/new samples/with_stat_area/{file_name}')
    stat_area_ids.append(cur_df.stat_area_id.unique())
```

```python
unique_stat_area_ids = np.unique(np.concatenate(stat_area_ids))
unique_stat_area_ids = unique_stat_area_ids[~np.isnan(unique_stat_area_ids)]
```

```python
np.save('../data/unique_stat_area_ids_in_cdr.npy',unique_stat_area_ids)
```
